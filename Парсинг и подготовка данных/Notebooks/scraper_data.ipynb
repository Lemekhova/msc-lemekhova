# Парсинг названий статей с сайта www.fontanka.ru.

Категории:
- Финансы
- Недвижимость
- Авто
- Спорт

```python
import requests
from bs4 import BeautifulSoup
import csv

output_file = "finances_60.csv"

with open(output_file, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["text", "label"])

    # перебираем страницы
    for page in range(1, 61):
        if page == 1:
            url = "https://www.fontanka.ru/finances/"
        else:
            url = f"https://www.fontanka.ru/finances/page-{page}/"

        print(f"Парсим {url}...")
        response = requests.get(url)
        html = response.content.decode("utf-8", errors="ignore")
        soup = BeautifulSoup(html, "html.parser")

        titles = soup.select("a[class^=header_]")
        subtitles = soup.select("span[class^=subtitle_]")

        for i, t in enumerate(titles):
            title = t.get_text(strip=True)
            subtitle = subtitles[i].get_text(strip=True) if i < len(subtitles) else ""
            text = f"{title}. {subtitle}" if subtitle else title
            writer.writerow([text, "finances"])
```

# Парсинг названий статей с сайта www.dp.ru.

Категории:
- Финансы
- Недвижимость
- Авто

```python
import time
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By


def parse_articles_with_selenium(url, label="realty", max_clicks=10):
    """
    Парсинг статей с возможностью:
    - менять URL
    - задавать label
    """

    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    driver = webdriver.Chrome(options=options)

    try:
        print(f"Загрузка страницы: {url}")
        driver.get(url)
        time.sleep(5)

        articles_data = []
        click_count = 0

        while click_count < max_clicks:
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            article_blocks = soup.find_all('div', class_='paragraphText')

            print(f"Клик #{click_count}, найдено: {len(article_blocks)}")

            for block in article_blocks:
                text_div = block.find('div', recursive=False)
                if text_div:
                    text = text_div.get_text(strip=True)

                    if text and text not in [a[0] for a in articles_data]:
                        articles_data.append((text, label))

            # нажимаем кнопку
            try:
                btn = driver.find_element(By.XPATH, "//span[contains(text(),'Загрузить еще')]")
                driver.execute_script("arguments[0].click();", btn)

                time.sleep(3)
                click_count += 1

            except:
                print("Кнопка закончилась")
                break

        print(f"Собрано: {len(articles_data)}")
        return articles_data

    finally:
        driver.quit()


def save_to_csv(data, filename="articles.csv"):
    with open(filename, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(["text", "label"])  # заголовки

        for row in data:
            writer.writerow(row)

    print(f"Сохранено в {filename}")
```

```python
# ЗАПУСК realty
url = "https://www.dp.ru/tag/%D0%9D%D0%B5%D0%B4%D0%B2%D0%B8%D0%B6%D0%B8%D0%BC%D0%BE%D1%81%D1%82%D1%8C"

articles = parse_articles_with_selenium(
    url=url,
    label="realty",
    max_clicks=200
)

save_to_csv(articles, "dp_realty_1.csv")
```
